One-sample t-test


A nutritionist claims the average sugar level in a new drink is 15 grams.
A sample of 8 bottles gives sugar values:

14, 16, 13, 17, 15, 14, 18, 16

Test whether the true mean sugar level is different from 15 grams using a one-sample t-test at 5% significance.


```python
import numpy as np
import scipy.stats as stats

# Create a dummy dataset of sugar level (grams) in 8 bottles
data = np.array([14, 16, 13, 17, 15, 14, 18, 16])
print("Data :", data)

# Define the null hypothesis
H0 = "The average sugar level in the drink is 15 grams."

# Define the alternative hypothesis
H1 = "The average sugar level in the drink is NOT equal to 15 grams."

print("Null Hypothesis (H0):", H0)
print("Alternative Hypothesis (H1):", H1)

# Calculate the test statistic (one-sample t-test)
t_stat, p_value = stats.ttest_1samp(data, 15)

# Print the results
print("\nTest statistic:", t_stat)
print("p-value:", p_value)

# Conclusion
if p_value < 0.05:
    print("Reject the null hypothesis. There is evidence that the mean sugar level is different from 15 grams.")
else:
    print("Fail to reject the null hypothesis. There is NO evidence that the mean sugar level is different from 15 grams.")
```

Two-Sample t-Test
A sports coach wants to check if two different training programs produce different average sprint times.

He takes two groups of athletes:

Group A (Program 1):
[12.1, 11.8, 12.5, 11.9, 12.3]

Group B (Program 2):
[11.5, 11.2, 11.7, 11.6, 11.4]

Test whether the mean sprint times of the two programs are different
(Use a two-tailed test, α = 0.05)

```python
import numpy as np
import scipy.stats as stats

# Two groups of sprint times
groupA = np.array([12.1, 11.8, 12.5, 11.9, 12.3])
groupB = np.array([11.5, 11.2, 11.7, 11.6, 11.4])

print("Group A:", groupA)
print("Group B:", groupB)

# Hypotheses
H0 = "The mean sprint time of Program A equals Program B."
H1 = "The mean sprint time of Program A is NOT equal to Program B."

print("\nNull Hypothesis (H0):", H0)
print("Alternative Hypothesis (H1):", H1)

# Means and SD
meanA = np.mean(groupA)
meanB = np.mean(groupB)
sdA = np.std(groupA, ddof=1)
sdB = np.std(groupB, ddof=1)

print("\nMean of Group A:", meanA)
print("Mean of Group B:", meanB)
print("SD of Group A:", sdA)
print("SD of Group B:", sdB)

# Two-sample t-test (independent, unequal variance)
t_stat, p_value = stats.ttest_ind(groupA, groupB, equal_var=False)

print("\nCalculated t-value:", t_stat)
print("p-value:", p_value)

# Degrees of freedom (Welch's)
df = len(groupA) + len(groupB) - 2
print("Degrees of Freedom (approx):", df)

# Critical t-value for 5% significance (two-tailed)
critical_t = stats.t.ppf(1 - 0.025, df)
print("Critical t-value (±):", critical_t)

# Decision
print("\n--- Decision ---")
print("Our t-value:", t_stat)
print("Critical t-value:", critical_t)

if abs(t_stat) > critical_t:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")

# Final conclusion
print("\nConclusion:")
if abs(t_stat) > critical_t:
    print("Reject H0: The two averages are different.")
else:
    print("Accept H0: The two averages are the same.")

```

Problem (1-sample Z-test)

A juice factory claims that the average fill of its bottles is 500 ml.
From past data, the population standard deviation (σ) is known to be 10 ml.

You take a sample of 40 bottles, and the sample mean comes out to 495 ml.

Check, at 5% significance, whether the machine is underfilling the bottles.



```python
import numpy as np
from math import sqrt
from scipy.stats import norm

# Given values
mu0 = 500          # Claimed population mean
sigma = 10         # Known population standard deviation
n = 40             # Sample size
sample_mean = 495  # Sample mean

print("Claimed population mean (mu0):", mu0)
print("Known population SD (sigma):", sigma)
print("Sample size (n):", n)
print("Sample mean (x̄):", sample_mean)

# Standard error
SE = sigma / np.sqrt(n)
print("\nStandard Error (SE):", SE)

# Z statistic
z_stat = (sample_mean - mu0) / SE
print("Calculated Z value:", z_stat)

# Significance level
alpha = 0.05

# Critical Z value for LEFT-tailed test (H1: mu < mu0)
critical_z = norm.ppf(alpha)  # because it's left-tailed
print("Critical Z value:", critical_z)

# p-value for left-tailed test
p_value = norm.cdf(z_stat)
print("p-value:", p_value)

print("\n--- Decision ---")
print("Our Z value:", z_stat)
print("Critical Z value:", critical_z)

if z_stat < critical_z:
    print("Reject H0 → The mean fill is LESS than 500 ml (underfilling).")
else:
    print("Accept H0 → The mean fill is NOT less than 500 ml (no underfilling).")
```

Two factories, A and B, produce light bulbs.

From past records:

Population standard deviation of lifetime from Factory A: σ₁ = 100 hours

Population standard deviation of lifetime from Factory B: σ₂ = 120 hours

You take samples:

From Factory A: n₁ = 50 bulbs, sample mean lifetime x̄₁ = 830 hours

From Factory B: n₂ = 60 bulbs, sample mean lifetime x̄₂ = 780 hours

At 5% significance, test whether the average lifetimes of bulbs from the two factories are different.

```python
import numpy as np
from math import sqrt
from scipy.stats import norm

# Given population standard deviations
sigma1 = 100   # Factory A
sigma2 = 120   # Factory B

# Given sample sizes
n1 = 50        # Sample from Factory A
n2 = 60        # Sample from Factory B

# Given sample means
mean1 = 830    # Mean lifetime from Factory A sample
mean2 = 780    # Mean lifetime from Factory B sample

print("Factory A: sigma1 =", sigma1, ", n1 =", n1, ", sample mean1 =", mean1)
print("Factory B: sigma2 =", sigma2, ", n2 =", n2, ", sample mean2 =", mean2)

# Hypotheses
H0 = "The mean lifetime of bulbs from Factory A equals Factory B. (mu1 = mu2)"
H1 = "The mean lifetime of bulbs from Factory A is NOT equal to Factory B. (mu1 != mu2)"

print("\nNull Hypothesis (H0):", H0)
print("Alternative Hypothesis (H1):", H1)

# Standard Error of (mean1 - mean2)
SE = sqrt((sigma1**2) / n1 + (sigma2**2) / n2)
print("\nStandard Error (SE) of (mean1 - mean2):", SE)

# Z statistic
z_stat = (mean1 - mean2) / SE
print("Calculated Z value:", z_stat)

# Significance level
alpha = 0.05

# Critical Z value for TWO-tailed test
critical_z = norm.ppf(1 - alpha/2)
print("Critical Z value (±):", critical_z)

# p-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z_stat)))
print("p-value:", p_value)

print("\n--- Decision ---")
print("Our Z value:", z_stat)
print("Critical Z value:", critical_z)

if abs(z_stat) > critical_z:
    print("Reject H0 → The two means are different.")
else:
    print("Accept H0 → The two means are the same.")

print("\nConclusion:")
if abs(z_stat) > critical_z:
    print("The average lifetimes of bulbs from Factory A and Factory B are DIFFERENT.")
else:
    print("The average lifetimes of bulbs from Factory A and Factory B are NOT different.")
```

chi - square
Problem (Goodness-of-Fit – very simple)

You roll a die 60 times and get:

Face 1 → 8 times

Face 2 → 10 times

Face 3 → 9 times

Face 4 → 12 times

Face 5 → 11 times

Face 6 → 10 times

Observed frequencies:
[8, 10, 9, 12, 11, 10]

If the die is fair, each face should appear equally often.

Total rolls = 60
Expected per face = 60 / 6 = 10

Expected frequencies:
[10, 10, 10, 10, 10, 10]

Question:

Is the die fair or not fair at 5% significance?

```python
import numpy as np
from scipy.stats import chisquare, chi2

# Observed frequencies for each face of a die (from 60 rolls)
observed = np.array([8, 10, 9, 12, 11, 10])
print("Observed frequencies :", observed)

# Expected frequencies if the die is fair (each face should appear 10 times)
expected = np.array([10, 10, 10, 10, 10, 10])
print("Expected frequencies :", expected)

# Hypotheses
H0 = "The die is fair. (All faces have equal probability)"
H1 = "The die is NOT fair. (At least one face has different probability)"

print("\nNull Hypothesis (H0):", H0)
print("Alternative Hypothesis (H1):", H1)

# Chi-square test (goodness-of-fit)
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("\nCalculated Chi-square value (χ²):", chi_stat)
print("p-value:", p_value)

# Degrees of freedom
df = len(observed) - 1
print("Degrees of Freedom (df):", df)

# Critical chi-square value at 5% significance (right-tailed test)
alpha = 0.05
critical_chi = chi2.ppf(1 - alpha, df)
print("Critical Chi-square value at 5% (χ²_critical):", critical_chi)

print("\n--- Decision ---")
print("Our χ² value:", chi_stat)
print("Critical χ² value:", critical_chi)

if chi_stat > critical_chi:
    print("Reject H0 → The die is NOT fair.")
else:
    print("Accept H0 → The die is fair.")

print("\nConclusion:")
if chi_stat > critical_chi:
    print("The observed frequencies are too different from expected → die is biased.")
else:
    print("The observed frequencies are close to expected → die behaves like a fair die.")
```
